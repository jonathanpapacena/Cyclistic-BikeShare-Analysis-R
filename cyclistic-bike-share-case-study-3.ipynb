#### I then performed some data manipulation tasks with the dplyr package from tidyverse.

#### I needed to see exactly how long each trip lasted so I created a column containing the difference between **ended_at** and **started_at**.

#### I wanted to know the day of the week that each trip was started, so that I can see patterns and differences in bike ride volume by day.

#### I then converted convert the new columns from character to factor class so that they can be chronologically ordered in the analysis, instead of alphabetically.

```r
# load the dplyr package to use the mutate function used for creating new columns
library(dplyr)

# add a calculated column of the ride length in minutes
# permanently changing the data frame by assigning the same name to this variable

bike_rides2023 <- 
  bike_rides2023 %>% 
  mutate(ride_length = difftime(ended_at,started_at, units = "mins"), .after = ended_at)

# add a column for the day of the week in which a bike trip begins
# convert the new column from character to factor class, in specified order
bike_rides2023 <-
  mutate(bike_rides2023, day_of_week = weekdays(started_at, abbreviate = TRUE), .after = started_at) %>% 
  mutate(day_of_week = factor(day_of_week, levels = c("Mon","Tue","Wed","Thu","Fri","Sat","Sun")))

# add a column for the month in which the bike trip begins
# convert the new column from character to factor class, in specified order
bike_rides2023 <-
  mutate(bike_rides2023, month = months(started_at, abbreviate = TRUE), .after = day_of_week) %>% 
  mutate(month = factor(month, levels = c("Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec")))

# the new and improved data frame
head(bike_rides2023)
```

#### After sorting the dataframe by ride_length in ascending order, I discovered many negative and zero values.

#### The negative values were a result of the start_date occuring after the end_date, which is impossible, and the zero values were calculated by the start date and end date being the same.

#### Since the negative and zero value ride lengths defy logic but more importantly can skew the data or lead to inaccurate findings, I removed them and created a new dataframe containing 5,718,608 rows, 1,269 less than before.

#### I created a new dataframe for this just incase I want to go back to the previous one.

```r
# sort by ride_length in ascending order to see that there are several negative values
bike_rides2023 %>% 
  arrange(ride_length)

# created a new dataframe with the negative and zero ride_length values removed by filtering them out with the filter() function
new_bike_rides2023 <-
  bike_rides2023 %>% 
  filter(ride_length > 0)

# I see my new row count and the correct order of my factor column variables
str(new_bike_rides2023)
```

### **Analyze**

### **4. A summary of my analysis**

#### Now that the dataset is ready for analysis, here is a summary of my exploratory analysis:

#### First, I will display the code that was used, followed by the observations made from the output.

```r
# Finding key summary statistics using the dplyr summarize() and group_by() functions

# Find the average ride_length of casual riders and members
new_bike_rides2023 %>% 
  group_by(member_casual) %>% 
  summarize(avg_ride_length = mean(ride_length))
```


#### The average ride length of casual riders is 28.25 minutes. This is more than double the average ride length of members, being 12.53 minutes.

```r
# Find the maximum ride length of casual riders and members
new_bike_rides2023 %>% 
  group_by(member_casual) %>% 
  summarize(max_ride_length = max(ride_length))
```

#### The maximum ride length of casual riders and overall, was 98489.07 minutes. That is 1641.49 hours or 68.4 days.

#### Members have a maximum ride length of 1559.67 minutes, or about 26 hours.

```r
# Find the average ride_length of casual riders and members by day_of_week
new_bike_rides2023 %>% 
  group_by(day_of_week, member_casual) %>% 
  summarize(avg_ride_length = mean(ride_length))
```

#### On average, casual riders spend more time on bike trips than members; about double the member's ride length during the week and nearly triple the ride length of members on the weekends.

#### For both, the average ride length remains fairly consistent throughout the week. Their average ride length slightly increases on the weekend.

```r
# Find the total amount of bike rides by casual riders and members on each day of the week for the calendar year
new_bike_rides2023 %>%
  group_by(day_of_week, member_casual) %>% 
  count(member_casual)
```

#### There are more total bike trips by members on every day of the week.

#### There is a sudden, significant increase in total casual rider bike trips on the weekends, with the most being on Saturday. Casuals seem to ride much more on the weekends than they do during the week.

#### For members, it is the opposite:

#### Most of their bike trips take place on weekdays, peaking on Thursday and then dropping on the weekends.

#### They have the fewest rides on Sunday.

```r
# Find the total amount of bike trips by members and casuals for each month of the year

print(new_bike_rides2023) %>%
  group_by(month, member_casual) %>%
  count(member_casual) %>% 
  print(n = 24)
```

#### Both members and casuals have the fewest trips from November to March.

#### The most bike trips happen from May to September.

#### **Members:**

#### The lowest amount of total trips is in February: 147,422

#### **Casuals:**
#### The peak amount of total casual's bike trips is in July: 331,252.

#### This visualization displays a comparison of casual rider and member usage by day of the week:

```r
# bar graph displaying bike trip count of members and casual riders for each day of the week in 2023.

ggplot(data = new_bike_rides2023) +
  geom_bar(mapping = aes(x = day_of_week, fill = member_casual), show.legend = FALSE) +
  geom_text(aes(x = day_of_week, y = ..count.., label = comma(..count..)), stat = "count", vjust = 1.5, size = 2.5) +
  labs(title = "Total Bike Rides per Day in 2023", x = "Day of Week", y = "Number of Bike Trips", subtitle = "Casual Rider Usage vs Member Usage") +
  theme(plot.title = element_text(hjust = 0.5), plot.subtitle = element_text(hjust = 0.5)) +
  facet_wrap(~member_casual) +
  scale_y_continuous(labels = comma)
```

#### **Key Findings:**

#### There is a sudden, significant increase in total casual rider bike trips on the weekends, with the most being on Saturday. Casuals seem to ride much more on the weekends than they do during the week.

#### For members, it is the opposite:

#### Most of their bike trips take place on weekdays, peaking on Thursday and then dropping on the weekends.

#### They have the fewest rides on Sunday.

#### **Inisight based on the data:** Members use the bikes more for commuting to work or school while casuals ride for liesure. This could explain: 

#### The member's shorter ride durations with higher volume during the week and lower volume on the weekend.

#### The casual's longer ride durations with lower volume during the week and higher on the weekends.
